# Time and Space Complexity Plot Exercise

Goal: To visualize and understand time/space complexity 

```python
import sys
import numpy as np 
import pandas as pd 
import matplotlib.pyplot as plt
import datetime
```

```python
print('Python Version:', sys.version[:3])
print('Date:',datetime.datetime.today().date())
```

## Sample

```python
x = [1,2,3]
y = [1,2,3]
plt.plot(x,y)
plt.ylabel('space/time')
plt.xlabel('n')
plt.title('sample')
plt.show()
```

## O(Log N)

```python
n = np.arange(1,1000,1)
y = np.log(n)
plt.plot(n,y)
plt.ylabel('space/time')
plt.xlabel('n')
plt.title('log n')
plt.show()
```

```python
# print some values 
list(zip(n,y))[:10]
```

## O(N Log N)

```python
n = np.arange(1,1000,1)
y = np.log(n) * n
plt.plot(n,y)
plt.ylabel('space/time')
plt.xlabel('n')
plt.title('n log n')
plt.show()
```

```python
# print some values 
list(zip(n,y))[:10]
```

## O(N^2)

```python
n = np.arange(1,10000,1)
y = n * n
plt.plot(n,y)
plt.ylabel('space/time')
plt.xlabel('n')
plt.title('n^2')
plt.show()
```

## Combine

```python
n = np.arange(0.001,100)
plt.plot(n,np.log(n),label='log n')
plt.plot(n,np.log(n)*n,label='n log n')
plt.plot(n,n,label='n')
plt.plot(n,n*n,label='n^2')
plt.ylabel('space/time')
plt.xlabel('n')
plt.title('graphs')
plt.legend()
plt.ylim(0,10)
plt.xlim(0,10)
plt.show()
```

```python
# compare values 
n = np.arange(1,100)

pd.DataFrame({
    'n':n
    ,'log n':np.log(n)
    ,'n log n':np.log(n)*n
    ,'n^2':n*n
             }).head(10)
```
